## Was sad that P(A|B) = 0.2, P(A|B') = 0.3, and P(B) = 0.8. What is P(A)?
***

```python
from matplotlib_venn import venn2
import matplotlib.pyplot as plt
```

**If P(A|B) = 0.2 and P(B) = 0.8, then P(A $\cap$ B) = P(A|B) x P(B) = 0.8 x 0.2 = 0.16**<br>
**If P(A|B') = 0.3 and P(B') = 1 - P(B) = 1 - 0.8 = 0.2, then P(A $\cap$ B') = P(A|B') x P(B') = 0.3 x 0.2 = 0.06**<br>
**But, P(A $\cap$ B') is the part that exist exclusively in A and not in B (P(A-B)).**<br>
**P(B-A) = P(B) - P(A $\cap$ B) = 0.8 - 0.16 = 0.64**

```python
venn2(subsets = (0.06, 0.64, 0.16))
plt.show()
```

**Finally, P(A) = P(A-B) + P(A $\cap$ B) = 0.06 + 0.16 = 0.22**
